# Functions

```python
def prepare_text(text):
  chars_to_remove = string.punctuation + "‘’“”…"
  prepared_text = ''.join([char for char in text if char not in chars_to_remove])
  prepared_text = prepared_text.lower()
  
  return prepared_text
```

```python
def tokenize_text(text):
  tokenized_text = nltk.word_tokenize(text)

  stopwords = nltk.corpus.stopwords.words('english')
  stopwords.extend(['video', 'images', 'image', 'tweets', 'tweet', 
                    'imagestweets', 'reports', 'report', 'photos', 
                    'photo', 'details', 'breaking', 'news', 'whoa', 
                    'wow', 'watch', 'factbox'])
  tokenized_text = [word for word in tokenized_text if word not in stopwords]

  # get pairs with word and part of speech (pos) extended symbol
  word_pos_pairs = nltk.pos_tag(tokenized_text)
  
  # only 4 main parts of speech for lemmatizer, selection based on first character of pos
  part_of_speech_for_lemma = {
    'J': 'a',       # adjective
    'N': 'n',       # noun
    'R': 'r',       # adverb
    'V': 'v'        # verb
  }
  lemmatized_words = []
  lemmatizer = nltk.WordNetLemmatizer()
  for word_pos_pair in word_pos_pairs:
    lemmatized_words.append(lemmatizer.lemmatize(
        word_pos_pair[0], 
        part_of_speech_for_lemma.get(word_pos_pair[1][0], 's')  # 's' otherwise
    ))
    
  return ' '.join(lemmatized_words)
```

```python
def transform_news(text):
  result = prepare_text(text)
  result = tokenize_text(result)
  return result
```

```python
def add_transformed_news(dataset, col_source, col_target):
  df = dataset.copy()
  df[col_target] = df[col_source].apply(lambda x: transform_news(x))
  return df
```

# Summary

Titles of news in four datasets were analysed, unnecessary columns were removed, as well as rows with nulls in title or label. Common label 'real' means that news are not fake. Titles were transformed by deletion of punctuation (extended for this project on the basis of data analysis), using lowercase, tokenization, deletion of stopwords (also extended) and lemmatization. Transformed titles were saved in new column 'news'. Then datasets were concatenated. It turned out that there were several empty news after transformation, so these rows were removed. Final dataset contained 139.662 news, 64.264 of which were fake and 75.398 were real. It was saved to csv file.
